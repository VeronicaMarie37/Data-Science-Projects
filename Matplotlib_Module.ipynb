# Matplotlib Module

###### Create dataset

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10, 12]
```

```python
###### plot line chart
```

```python
plt.plot(x, y)
plt.show()
```

###### Label axes of the plot

```python
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

###### Add title to the plot

```python
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("Relation between input and output values")


plt.show()
```

###### Change the style of the plot

```python
plt.plot(x, y, linestyle='--' , color='red')

plt.show()
```

```python
###### Change the color of the plot
```

```python
#  legends
plt.plot(x, y, label='Line-1')

plt.show()
```

```python
#  legends
plt.plot(x, y, label='Line-1')
plt.legend()
plt.show()
```

###### Scattered plot

```python
# scatter() function is used to plot scattered
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel("Number of hours")
plt.ylabel("Marks achieved")
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel("Number of hours")
plt.ylabel("Mathematics marks")
plt.title("Relation between number of hours vs marks")
plt.show()

```

```python
plt.scatter(x, y, c='r')
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10, 12]
z = [2, 3, 4, 2, 5, 1]
```

```python
plt.plot(x, z)
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, z, label='Mathematics')
plt.plot(x, y, label = 'Science')
plt.xlabel("Hoours")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
plt.plot(x,y, label='Mathematics')
plt.scatter(x, z, label='Science', c='r')

plt.legend()

plt.show()
```

###### Bar plot

```python
plt.bar(x, z)

plt.show()
```

```python
plt.bar(x, y)
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

###### Pie Plot

```python
plt.pie(y, labels=x)
plt.show()
```

```python
label = ['Apple', 'Orange', 'Mango']
price = [300, 240, 360]
```

```python
plt.pie(price, labels = label)
plt.show()
```

```python
plt.pie(price, labels = label, autopct = '%1.1f%%')
plt.show()
```

```python
my_explode = [0.09, 0.05, 0]
plt.pie(price, labels = label, autopct = '%1.1f%%', explode=my_explode)
plt.show()
```

###### Subplots

```python
plt.subplot(2, 2, 1)
plt.plot(x,y)

plt.subplot(2, 2, 2)
plt.scatter(x, y)

plt.subplot(2, 2, 3)
plt.bar(x, y)

plt.subplot(2, 2, 4)
plt.pie(price, labels = label)

plt.show()
```
