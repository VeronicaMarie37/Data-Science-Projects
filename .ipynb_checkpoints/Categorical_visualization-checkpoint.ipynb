```python
import pandas as pd

data = pd.read_csv("house.csv")
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
# visualization
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=data, x='number_of_rooms')
```

```python
sns.countplot(data=data, x='number_of_rooms', palette = 'viridis')
plt.title("Categorical values")
```

```python
sns.countplot(data=data, x='number_of_rooms', hue='floor')
```

```python
#  violin
```

```python
sns.violinplot(data['price'])
```

```python
sns.violinplot(data['number_of_rooms'])
```

```python
sns.violinplot(data=data, x='number_of_rooms', y='price')
```

```python
sns.violinplot(data=data, x='price', y='number_of_rooms')
```

```python
sns.boxplot(data)
```

```python
sns.boxplot(data['number_of_rooms'])
```

```python
sns.boxplot(data= data, x='number_of_rooms', y='price')
```

```python
# swarm plot
```

```python
sns.swarmplot(data=data, x='number_of_rooms', y='area')
```

```python
sns.swarmplot(data=data, x='number_of_rooms', y='floor')
```

```python
sns.stripplot(data=data, x='number_of_rooms', y='area')
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot=True)
```
