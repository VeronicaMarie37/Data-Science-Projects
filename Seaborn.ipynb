# Seaborn Module

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

###### Scatter plots

```python
sns.scatterplot(data = tips, x='tip', y='total_bill')
```

```python
sns.scatterplot(data = tips, x='tip', y='total_bill')
plt.title("Bill vs Tips")
plt.show()
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
```

###### Line plots

```python
sns.lineplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day')
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day')
plt.title("Tips vs Bill Vs days")

plt.show()
```

###### Bar plots

```python
sns.barplot(data=tips, x='day', y='tip')
```

```python
sns.barplot(data=tips, x='time', y='tip')
```

```python
sns.barplot(data=tips, x='day', y='tip', hue='sex')
```

```python
sns.barplot(data=tips, x='time', y='tip', hue='sex')
```

###### Histogram plots

```python
sns.histplot(tips['tip'])
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.histplot(tips['total_bill'] , kde=True)
```

###### Box Plots


```python
tips.head()
```

```python
tips.info()
```

```python
sns.boxplot(data=tips)
```

```python
sns.boxplot(data=tips, x='day', y='tip')
```

```python
sns.boxplot(data=tips, x='day', y='tip', hue='sex')
```

###### Voilin plots

```python
sns.violinplot(data=tips)
```

```python
sns.violinplot(x='day', y='tip', data=tips)
```

###### Heatmaps

```python
sns.heatmap(tips.corr(), annot=True, cmap='Oranges')
```

###### Pair plots

```python
sns.pairplot(data=tips)
```

```python
sns.pairplot(data=tips, hue='sex')
```

```python
sns.pairplot(data=tips, hue='sex', kind='kde')
```
