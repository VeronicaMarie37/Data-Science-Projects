```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("house.csv")
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#line plot
plt.plot(df.area.index,df.area)
plt.show()
```

```python
plt.plot(df.area.index,df.area, linestyle="--", color="purple", label="area")
plt.xlabel("showing index values")
plt.ylabel("showing area values")
plt.title("index vs area")
plt.legend()
plt.show()
```

```python
#scattered plot
plt.scatter(df.price, df.area, c="green")
plt.xlabel("price")
plt.ylabel("area")
plt.title("price vs area")
plt.show()
```

```python
plt.plot(df.price.index, df.price, label="price")
plt.plot(df.area.index,df.area, label="area")
plt.xlabel("area")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
plt.scatter(df.price.index, df.price, label="price")
plt.scatter(df.area.index,df.area, label="area")
plt.xlabel("area")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
plt.scatter(df.area.index,df.area, label="area")
plt.scatter(df.latitude.index, df.latitude, label="latitude")
plt.xlabel("area")
plt.ylabel("latitude")
plt.legend()
plt.show()
```

```python
plt.scatter(df.longitude.index,df.longitude, label="longitude")
plt.scatter(df.latitude.index, df.latitude, label="latitude")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.legend()
plt.show()
```

```python
plt.plot(df.longitude.index,df.longitude, label="longitude")
plt.plot(df.latitude.index, df.latitude, label="latitude")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.legend()
plt.show()
```

```python
plt.bar(df.floor,df.floor, label="floor")
plt.xlabel("floor")
plt.ylabel("floor 2")
plt.legend()
plt.show()
```

```python
plt.hist(df.price, bins=100, label="price")
plt.xlabel("price")
plt.ylabel("price1")
plt.legend()
plt.show()
```

```python
plt.pie(df.floor.value_counts(), labels=df.floor.unique(), autopct="%1.1f%%")
plt.show()
```

```python
df.number_of_rooms.unique()
```

```python
my_explode =[0.09, 0.05, 0.04, 0.02, 0,0]
plt.pie(df.number_of_rooms.value_counts(), labels=df.number_of_rooms.unique(), autopct="%1.1f%%", explode = my_explode)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(df.area.index,df.area, linestyle="--", color="purple", label="area")

plt.subplot(2,2,2)
plt.scatter(df.price.index, df.price, label="price")

plt.subplot(2,2,3)
plt.bar(df.floor,df.floor, label="floor")

plt.subplot(2,2,4)
plt.pie(df.floor.value_counts(), labels=df.floor.unique(), autopct="%1.1f%%")

plt.savefig("subplot_display.jpeg")
```
